Цель: на данной практической работе необходимо использовать датасет с прошлой практической работы и визуализировать данные в виде графиков.Необходимо используя библиотеки сгенерировать следующие графики: линейный(мин. 3шт), столбчатый(мин. 3шт), круговая (мин. 2шт), рассеяния(мин. 2шт), гистограмма(мин. 3шт) и описать их.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('new_sw_dataset.csv')
data = pd.DataFrame(data)
```

Импортируем библиотеки для работы с нашим датасетом и визуализации данных. Создадим 3 линейных графика.

```python
plt.plot(data['height'], marker='o', color='green')
plt.title('Рост персонажей Звездных войн', fontweight='bold')
plt.xlabel('Персонажи', fontweight='bold')
plt.ylabel('Рост (см)', fontweight='bold')
plt.tight_layout()
plt.show()
```

Удалось создать линейный график №1 Рост персонажей Звездных войн. Из данного графика можно сделать вывод о том, что большинство персонажей из Звездных войн ростом от 150 до 200 снатиметров, а самый высокий имеет рост более 250 см. Самый низкий - менее 75 см.

```python
plt.plot(data['name'], data['age'], marker='o', color='blue')
plt.title('Возраст персонажей Звездных войн', fontweight='bold')
plt.xlabel('Персонажи', fontweight='bold')
plt.ylabel('Возраст (лет)', fontweight='bold')
plt.xticks(rotation=90, ha='left', fontsize=7)
plt.tight_layout()
plt.show()
```

Удалось создать линейный график №2 Возраст персонажей Звездных войн. Из данного графика можно сделать вывод о том, что большинство персонажей из Звездных войн возрастом не более 100 лет, а один из них старше 800 лет.

```python
plt.plot(data['weight'], marker='o', color='red')
plt.title('Вес персонажей из Звездных войн', fontweight='bold')
plt.ylabel('Вес персонажа (кг)',fontweight='bold')
plt.xlabel('Персонажи', fontweight='bold')
plt.xticks(rotation=90, ha='left', fontsize=6)
plt.tight_layout()
plt.show()
```

Удалось создать линейный график №3 Вес персонажей Звездных войн. Средний вес персонажей не превышает 100кг, вес самого тяжелого персонажа составляет около 1400кг, а самого легкого - около 60.
Создалим 3 столбчатых диаграммы.

```python
plt.bar(data['name'], data['films'].array, color='cyan')
plt.title('Количество фильмов для каждого персонажа в Звездных войнах', fontweight='bold')
plt.xlabel('Персонажи', fontweight='bold')
plt.ylabel('Количество фильмов', fontweight='bold')
plt.xticks(rotation=90, ha='right', fontsize=7)
plt.yticks(fontsize=7)
plt.show()
```

Удалось создать столбчатую диаграмму №1, она показывает в каких фильмах принимал участие тот или иной персонаж.
Подавляющее количество персонажей принимало учатие хотя-бы в 1 фильме, а чаще всего в большенстве из картин.

```python
plt.bar(data['name'], data['height'], color='purple')
plt.title('Рост каждого персонажа в Звездных войнах', fontweight='bold')
plt.xlabel('Персонажи', fontweight='bold')
plt.ylabel('Рост(см)', fontweight='bold')
plt.xticks(rotation=90, ha='right', fontsize=5)
plt.yticks(fontsize=7)
plt.show()
```

Удалось создать столбчатую диаграмму №2, она показывает рост каждого из персонажей Звездных войн. По данным из диаграммы видно, что средний рост персонажа из Звездных войн составляет приблизительно 175 сантиметров.

```python
plt.bar(data['name'], data['weight'], color='orange')
plt.title('Вес каждого персонажа в Звездных войнах', fontweight='bold')
plt.ylabel('Персонажи', fontweight='bold')
plt.xlabel('Вес(кг)', fontweight='bold')
plt.xticks(rotation=90, ha='right', fontsize=5)
plt.yticks(fontsize=7)
plt.show()
```

Удалось создать столбчатую диаграмму №3, она показывает вес каждого из персонажей Звездных войн.
Давайте создадим 2 круговых диаграммы. Из данной диаграммы следует, что у большинства персонажей вес находится в районе 100кг, вес самого тяжелого персонажа составляет около 1400кг.

```python
drop_data = data.dropna()
plt.pie(drop_data['height'], labels=drop_data['name'])
plt.title('Сравнение роста некоторых из персонажей в Звездных войнах')
plt.show()
```

Данная диаграмма показывает сравнение роста некоторых из персонажей Звездных войн. Смотря на данную диаграмму можно сделать вывод, что самым высоким из данного сравнения является Чубака. Остальные же примерно равного роста.

```python
drop_data = data.dropna()
plt.pie(drop_data['age'], labels=drop_data['name'], autopct='%1.1f%%')
plt.title('Доля возраста некоторых из персонажей в Звездных войнах')
plt.show()
```

Данная диаграмма показывает сравнение возраста в процентном соотношении некоторых из персонажей Звездных войн.
Давайте создадим 3 диаграммы рассеяния. Смотря на данную диаграмму можно сделать вывод, что Чубака старше всех выбранных персонажей, а Люк - самый молодой.

```python
x = data['weight']
y = data['height']
plt.scatter(x, y, color='blue')
plt.title('Соотношение веса и роста персонажей', fontweight='bold')
plt.xlabel('Вес(кг)', fontweight='bold')
plt.ylabel('Рост(см)', fontweight='bold')
plt.show()
```

Данная диаграмма рассеяния показывает соотношения по весу и росту персонажа Звездных войн. В выводе можно сказать, что чаще всего персонажи имеют рост от 150 до 200 и вес не превышающий 100кг.

```python
x = data['eye_color']
y = data['height']
plt.scatter(x, y, color='yellow')
plt.title('Цвет глаз у персонажей различного роста', fontweight='bold')
plt.xlabel('Цвет глаз', fontweight='bold')
plt.ylabel('Рост', fontweight='bold')
plt.xticks(rotation=90, ha='right', fontsize=8)
plt.show()
```

Данная диаграмма рассеяния показывает соотношения по цвету глаз и росту персонажа Звездных войн. По данным из диаграммы мы видим, что самый частовстречающийся цвет глаз - карий, а рост до 190 метров. Это говороит о том, что чаще всего персонажми выступают люди.
Давайте создадим 3 гистограммы.

```python
plt.hist(data['height'], bins=15, color='skyblue', edgecolor='black')
plt.title('Распределение роста персонажей в Звездных войнах')
plt.xlabel('Рост (см)')
plt.ylabel('Частота')
plt.show()
```

Данная гистограмма показывает популярность (частоту) роста персонажей Звездных войн. По данной гистограмме можно сделать вывод, что самый распространенный рост ~ от 160см до 2м.

```python
plt.hist(data['age'], bins=20, color='lightcoral', edgecolor='black')
plt.title('Распределение возраста персонажей в Звездных войнах')
plt.xlabel('Возраст')
plt.ylabel('Частота')
plt.show()
```

Данная гистограмма показывает распределение (частоту) возраста персонажей Звездных войн. По данным графика можно сделать вывод, что подавляющая часть персонажей чаще всего имеют возраст до 100 лет.

```python
plt.hist(data['eye_color'].array, color='lightgreen', edgecolor='black')
plt.title('Редкость цветов глаз у персонажей в Звездных войнах')
plt.xlabel('Цвет глаз')
plt.ylabel('Частота')
plt.xticks(rotation=90, ha='left', fontsize=8)
plt.show()
```

Данная гистограмма показывает популярность (редкость) цветов глаз персонажей Звездных войн. Исходя из полученного грфика можно сказать, что самый редкий цвет глаз - зелёный и жёлтый, а самый распространенный - коричневый.
